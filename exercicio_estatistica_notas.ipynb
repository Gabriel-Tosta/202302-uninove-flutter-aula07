Duas salas de aula tiveram as seguintes notas:

Notas1 =(8.7, 8.2, 6.1, 5.5, 7.1, 6.6, 7.7, 6.8,8.4, 6.6)<BR>
Notas2 =(10.0, 7.2, 7.1, 8.6, 8.5,7.8,6.5, 8.7, 2.0, 5.2, 7.2)

Descreva  o perfil de cada classe através da estatística descritiva.

media, mediana, desvio padrão, minimo , máximo

utilizar biblioteca NUMPY.

```python
import numpy as np
Notas1 =(8.7, 8.2, 6.1, 5.5, 7.1, 6.6, 7.7, 6.8,8.4, 6.6)
Notas2 =(10.0, 7.2, 7.1, 8.6, 8.5,7.8,6.5, 8.7, 2.0, 5.2, 7.2)
```

```python
# Calculate and print descriptive statistics for Notas1
print("Estatística descritiva para Notas1:")
print(f"Média: {np.mean(Notas1):.2f}")
print(f"Mediana:{ np.median(Notas1):.2f}")
print(f"Desvio Padrão:{ np.std(Notas1):.2f}")
print("Mínimo:", np.min(Notas1))
print("Máximo:", np.max(Notas1))

print("\nEstatística descritiva para Notas2:")
print(f"Média: {np.mean(Notas2):.2f}")
print(f"Mediana:{ np.median(Notas2):.2f}")
print(f"Desvio Padrão:{np.std(Notas2):.2f}")
print("Mínimo:", np.min(Notas2))
print("Máximo:", np.max(Notas2))
```

Pela estatistica descritiva:<br>
Podemos afirmar que a media de notas dos alunos da sala 1 e sala 2 são iguais.<br>
Mas em relação ao desvio padrão os alunos da sala 2 tem as notas mais dispersas, ou seja mais longe da media, e a sala 1 as notas estão mais proximas da media.

import numpy as np

# Notas das turmas
Notas1 = np.array([8.7, 8.2, 6.1, 5.5, 7.1, 6.6, 7.7, 6.8, 8.4, 6.6])
Notas2 = np.array([10.0, 7.2, 7.1, 8.6, 8.5, 7.8, 6.5, 8.7, 2.0, 5.2, 7.2])

# Estatísticas descritivas para Notas1
media1 = np.mean(Notas1)
mediana1 = np.median(Notas1)
desvio_padrao1 = np.std(Notas1)
minimo1 = np.min(Notas1)
maximo1 = np.max(Notas1)

# Estatísticas descritivas para Notas2
media2 = np.mean(Notas2)
mediana2 = np.median(Notas2)
desvio_padrao2 = np.std(Notas2)
minimo2 = np.min(Notas2)
maximo2 = np.max(Notas2)

print("\nEstatística descritiva para Notas1:")
print(f"Média: {media1:.2f}")
print(f"Mediana: {mediana1:.2f}")
print(f"Desvio Padrão: {desvio_padrao1:.2f}")
print(f"Mínimo: {minimo1}")
print(f"Máximo: {maximo1}")

print("\nEstatística descritiva para Notas2:")
print(f"Média: {media2:.2f}")
print(f"Mediana: {mediana2:.2f}")
print(f"Desvio Padrão: {desvio_padrao2:.2f}")
print(f"Mínimo: {minimo2}")
print(f"Máximo: {maximo2}")
