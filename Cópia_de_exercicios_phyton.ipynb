calcule a media de 3 notas de um aluno


se a média for maior ou igual a 6 : "Aluno Aprovado"


se for menor que 6 : "Aluno reprovado"

```python
def calcular_media_ponderada():
    print("--- Calculadora de Média Ponderada ---")

    try:

        nota1 = float(input("Digite a primeira nota: "))
        peso1 = float(input("Digite o peso da primeira nota: "))

        nota2 = float(input("\nDigite a segunda nota: "))
        peso2 = float(input("Digite o peso da segunda nota: "))

        nota3 = float(input("\nDigite a terceira nota: "))
        peso3 = float(input("Digite o peso da terceira nota: "))

    except ValueError:
        print("\nErro: Por favor, digite apenas números válidos para notas e pesos.")
        return


    soma_produtos = (nota1 * peso1) + (nota2 * peso2) + (nota3 * peso3)


    soma_pesos = peso1 + peso2 + peso3


    if soma_pesos == 0:
        print("\nErro: A soma dos pesos não pode ser zero.")
        return


    media_ponderada = soma_produtos / soma_pesos


    print(f"\nA soma dos pesos é: {soma_pesos}")
    print(f"A Média Ponderada do aluno é: {media_ponderada:.2f}")

    if media_ponderada >= 6:
        print("RESULTADO: Aluno Aprovado ")
    else:
        print("RESULTADO: Aluno reprovado ")


if __name__ == "__main__":
    calcular_media_ponderada()
```

Calcular o indice IMC

O Índice de Massa Corporal (IMC) é calculado dividindo o peso em quilogramas pela altura ao quadrado em metros. A fórmula é: IMC = peso / (altura x altura).

A tabela do IMC (Índice de Massa Corporal) relaciona o valor do IMC com a classificação de peso.


IMC Classificação <br>


Menor que 18,5 - Abaixo do peso <br>
entre 18,5–24,9 - Peso normal<br>
entre 25,0–29,9 - Sobrepeso<br>
entre 30,0–34,9 - Obesidade grau I<br>
entre 35,0–39,9 - Obesidade grau II<br>
entre 40,0 e acima - Obesidade grau III ou mórbida <br>

```python
def calcular_imc():
    print("--- Calculadora de Índice de Massa Corporal (IMC) ---")


    try:

        peso = float(input("Digite o seu peso em quilogramas (ex: 70.5): "))

        altura = float(input("Digite a sua altura em metros (ex: 1.75): "))

    except ValueError:
        print("\nErro: Por favor, digite valores numéricos válidos para peso e altura.")
        return


    if peso <= 0 or altura <= 0:
        print("\nErro: O peso e a altura devem ser valores positivos.")
        return


    imc = peso / (altura ** 2)


    print(f"\nSeu IMC é: **{imc:.2f}**")



    classificacao = ""

    if imc < 18.5:
        classificacao = "Abaixo do peso"
    elif imc <= 24.9:
        classificacao = "Peso normal"
    elif imc <= 29.9:
        classificacao = "Sobrepeso"
    elif imc <= 34.9:
        classificacao = "Obesidade grau I"
    elif imc <= 39.9:
        classificacao = "Obesidade grau II"
    else:
        classificacao = "Obesidade grau III ou mórbida"


    print(f"Sua classificação de peso é: **{classificacao}**")


if __name__ == "__main__":
    calcular_imc()
```

Faça um script para uma loja de tintas. <br>
O script deverá pedir o tamanho em metros quadrados da área a ser pintada.<br>

Considere que a cobertura da tinta é de 1 litro para cada 6 metros quadrados e que a tinta é vendida em latas de 18 litros,
que custam R 80,00 ou em galões de 3,6 litros, que custam R 25,00.<br>

Informe ao usuário as quantidades de tinta a serem compradas e os respectivos preços em 3 situações:<br>

comprar apenas latas de 18 litros;<br>

comprar apenas galões de 3,6 litros;<br>

misturar latas e galões, de forma que o preço seja o menor. Acrescente 10% de folga e sempre arredonde os valores para cima,
isto é, considere latas cheias.<br>


```python
import math


COBERTURA_POR_LITRO = 6.0
FOLGA_PERCENTUAL = 0.10


LATA_VOLUME = 18.0
LATA_PRECO = 80.00


GALAO_VOLUME = 3.6
GALAO_PRECO = 25.00

def calcular_compra_tintas():
    print("--- Calculadora de Tinta para Área ---")


    try:
        area_m2 = float(input("Digite o tamanho da área a ser pintada (em metros quadrados): "))
        if area_m2 <= 0:
            print("Erro: A área deve ser um valor positivo.")
            return
    except ValueError:
        print("Erro: Por favor, digite um valor numérico válido para a área.")
        return




    litros_necessarios_base = area_m2 / COBERTURA_POR_LITRO


    litros_necessarios_com_folga = litros_necessarios_base * (1 + FOLGA_PERCENTUAL)

    print(f"\n--- Análise de Necessidade ---")
    print(f"Litros necessários (Base): {litros_necessarios_base:.2f} L")
    print(f"Litros necessários (c/ 10% de Folga): {litros_necessarios_com_folga:.2f} L")
    print("-----------------------------\n")




    qtd_latas_A = math.ceil(litros_necessarios_com_folga / LATA_VOLUME)
    preco_total_A = qtd_latas_A * LATA_PRECO


    qtd_galao_B = math.ceil(litros_necessarios_com_folga / GALAO_VOLUME)
    preco_total_B = qtd_galao_B * GALAO_PRECO


    qtd_latas_C = math.floor(litros_necessarios_com_folga / LATA_VOLUME)


    litros_restantes = litros_necessarios_com_folga - (qtd_latas_C * LATA_VOLUME)


    qtd_galao_C = math.ceil(litros_restantes / GALAO_VOLUME)


    preco_latas_C = qtd_latas_C * LATA_PRECO
    preco_galao_C = qtd_galao_C * GALAO_PRECO
    preco_total_C = preco_latas_C + preco_galao_C



    print("###################################################")
    print("## RESULTADOS DA COMPRA DE TINTAS (c/ 10% Folga) ##")
    print("###################################################\n")


    print("Situação A: Comprar APENAS LATAS (18L)")
    print(f"  -> Quantidade de Latas (18L): {qtd_latas_A}")
    print(f"  -> Preço Total: R$ {preco_total_A:.2f}")
    print("-" * 50)


    print("Situação B: Comprar APENAS GALÕES (3,6L)")
    print(f"  -> Quantidade de Galões (3,6L): {qtd_galao_B}")
    print(f"  -> Preço Total: R$ {preco_total_B:.2f}")
    print("-" * 50)


    print("Situação C: Mistura para o MENOR PREÇO")
    print(f"  -> Quantidade de Latas (18L): {qtd_latas_C}")
    print(f"  -> Quantidade de Galões (3,6L): {qtd_galao_C}")
    print(f"  -> Preço Total: R$ {preco_total_C:.2f}")
    print("-" * 50)


    melhor_preco = min(preco_total_A, preco_total_B, preco_total_C)

    if melhor_preco == preco_total_C:
        print("\n✅ A melhor opção de compra é a Mistura (Situação C).")
    elif melhor_preco == preco_total_A:
        print("\n A melhor opção de compra é APENAS Latas (Situação A).")
    else:
        print("\n A melhor opção de compra é APENAS Galões (Situação B).")



if __name__ == "__main__":
    calcular_compra_tintas()


```

#Aplicativo de calculo de quantidade de caixas de piso
Uma loja de revestimentos ( pisos) necessita de um aplicativo que calcule o numero de caixas para cobertura de uma area fornecida pelo cliente. <br>
entradas:<br>
- largura e comprimento <br>
- dimensões do piso ( fornecido por uma tabela ) <br>
- opção: de piso liso ou desenhado <br>
saidas:<br>
quantidade de caixas exatas <br>
quantidade de pisos de folga ( caixas ) ( 10% e 5% )<br>
#************************
iniciar a criação do modulo phyton de calculo com:<br>
piso 30 x 30 cm<br>
piso liso<br>
faça um estudo sobre a quantidade de pisos calculado pela area total e pela quantidade de pisos -comprimento e quantidade de pisos-largura
#*******


```python
import math

def calcular_caixas_piso(largura_ambiente_m, comprimento_ambiente_m, dimensoes_piso_cm):




    area_ambiente_m2 = largura_ambiente_m * comprimento_ambiente_m


    largura_piso_m = dimensoes_piso_cm[0] / 100
    comprimento_piso_m = dimensoes_piso_cm[1] / 100


    area_piso_m2 = largura_piso_m * comprimento_piso_m


    area_por_caixa_m2 = pecas_por_caixa * area_piso_m2

    print(f"--- Detalhes do Cálculo ---")
    print(f"Área do ambiente: {area_ambiente_m2:.2f} m²")
    print(f"Área por peça de piso: {area_piso_m2:.2f} m²")
    print(f"Área por caixa ({pecas_por_caixa} peças): {area_por_caixa_m2:.2f} m²")
    print("---------------------------\n")





    qtd_total_pisos_area = math.ceil(area_ambiente_m2 / area_piso_m2)


    qtd_caixas_area = math.ceil(qtd_total_pisos_area / pecas_por_caixa)



    pecas_largura = math.ceil(largura_ambiente_m / largura_piso_m)

    pecas_comprimento = math.ceil(comprimento_ambiente_m / comprimento_piso_m)

    qtd_total_pisos_pecas = pecas_largura * pecas_comprimento

    qtd_caixas_pecas = math.ceil(qtd_total_pisos_pecas / pecas_por_caixa)


    qtd_caixas_base = max(qtd_caixas_area, qtd_caixas_pecas)



    folga_10_percent = 0.10
    qtd_caixas_10_folga = math.ceil(qtd_caixas_base * (1 + folga_10_percent))

    folga_5_percent = 0.05
    qtd_caixas_5_folga = math.ceil(qtd_caixas_base * (1 + folga_5_percent))


    resultados = {
        "metodo_area_total": {
            "qtd_pecas": qtd_total_pisos_area,
            "qtd_caixas_exatas": qtd_caixas_area,
        },
        "metodo_por_pecas": {
            "pecas_largura": pecas_largura,
            "pecas_comprimento": pecas_comprimento,
            "qtd_pecas": qtd_total_pisos_pecas,
            "qtd_caixas_exatas": qtd_caixas_pecas,
        },
        "resultado_final": {
            "qtd_caixas_base": qtd_caixas_base,
            "qtd_caixas_folga_10_pct": qtd_caixas_10_folga,
            "qtd_caixas_folga_5_pct": qtd_caixas_5_folga,
        }
    }

    return resultados


largura_cliente = 4.0
comprimento_cliente = 5.0 #

dimensoes_piso = (30, 30)

resultado_final = calcular_caixas_piso(largura_cliente, comprimento_cliente, dimensoes_piso)

print("\n#######################################################")
print("## RESULTADO DO CÁLCULO DE PISOS (30x30 cm) ##")
print("#######################################################")

print("\n### 1. Detalhes dos Métodos de Cálculo ###")
print(f"Cálculo Base (Área Total): {resultado_final['metodo_area_total']['qtd_caixas_exatas']} caixas")
print(f"Cálculo Base (Por Peças LxA): {resultado_final['metodo_por_pecas']['qtd_caixas_exatas']} caixas")
print(f"   -> Peças na Largura: {resultado_final['metodo_por_pecas']['pecas_largura']}")
print(f"   -> Peças no Comprimento: {resultado_final['metodo_por_pecas']['pecas_comprimento']}")
print(f"** Base Final de Caixas Necessárias (Maior Valor): {resultado_final['resultado_final']['qtd_caixas_base']} caixas **")

print("\n### 2. Quantidade de Caixas com Folga ###")
print(f"Quantidade de Caixas EXATAS (Sem Folga): {resultado_final['resultado_final']['qtd_caixas_base']}")
print(f"Quantidade de Caixas com FOLGA de 5%: {resultado_final['resultado_final']['qtd_caixas_folga_5_pct']}")
print(f"Quantidade de Caixas com FOLGA de 10%: {resultado_final['resultado_final']['qtd_caixas_folga_10_pct']}")
```
